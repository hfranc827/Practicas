```python
#Scatter plot
import matplotlib.pyplot as plt
x = [5,7,9,2,12,17]
Y = [99,87,56,97,46,78]
plt.scatter(x,Y)
plt.title('Scatter Plot')
plt.show()
```

```python
#Bar Chart
import matplotlib.pyplot as plt
names = ["A","B","C","D"]
values = [10,24,36,18]
plt.bar(names,values)
plt.title('Bar Chart')
plt.xlabel('Names')
plt.ylabel('Values')
plt.show()
```

```python
#Barra Horizontal
import matplotlib.pyplot as plt
names = ["A","B","C","D"]
values = [10,24,36,18]
plt.barh(names,values,color='orange')
plt.title("Barra Horizontal")
plt.show()
```

```python
#Multiple lineas
import matplotlib.pyplot as plt
x = [1,2,3,4]
y1 = [10,20,30,40]
y2 = [5,13,17,20]
plt.plot(x,y1, label ="linea1" , color = "blue")
plt.plot(x,y2, label ="linea2" , color = "red")
plt.title("Linas Multiples")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ["Python","Java","C++","JavaScript"]
tamanho = [45,25,15,15]
plt.pie(tamanho, labels = labels , autopct="%1.1f%%",startangle=90)
plt.title("Grafico Circular")
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [5,15,12,17,15,26,25,39,30]
plt.hist(data,bins=5 , color="skyblue",edgecolor="black")
plt.title("Histograma")
plt.show()
```
