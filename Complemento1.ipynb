```python
#Area de un triangulo
base = float(input("Ingrese la base del triangulo: "))
altura = float(input("Ingrese la altura del triangulo: "))
area = (base * altura)/2
print(f"El area del triangulo de base: {base} y altura: {altura} es -> {area}")
```

```python
#Convertir de kilometros a millas
kilometros = float(input("Ingrese la distancia en kilometros: "))
conversion_factor = 0.621371
millas = kilometros * conversion_factor
print(f"{kilometros} kilometros son -> {millas} millas")
```

```python
#Convertir Celsius a Farenheit
celsius = float(input("Ingrese la temperatura en celsius: "))
fahrenehit = (celsius * 9/5) + 32
print(f"{celsius} celsius son -> {fahrenehit} fahrenheit")
```

```python
#resolver ecuacion cuadratica
# a,b,c numeros reales
# a =! 0
import math
a = float(input("Ingrese el valor de a: "))
b = float(input("Ingrese el valor de b: "))
c = float(input("Ingrese el valor de c: "))
discriminante = b**2 - 4*a*c
if discriminante > 0:
  x1 = (-b + math.sqrt(discriminante))/(2*a)
  x2 = (-b - math.sqrt(discriminante))/(2*a)
  print(f"Las soluciones son: x1 = {x1} y x2 = {x2}")
elif discriminante == 0:
  x = -b/(2*a)
  print(f"La solucion es: x = {x}")
else:
  print("No tiene solucion real")
```

```python
#Factorial
num = int(input("Ingrese un numero: "))
factorial = 1
if num < 0:
  print("No existe el factorial de un numero negativo")
elif num == 0:
  print("El factorial de 0 es 1")
else:
  for i in range(1, num + 1):
    factorial = factorial * i
  print(f"El factorial de {num} es {factorial}")
```
