## INTRODUCCION A PANDAS

```python
#Create Dataframes
import pandas as pd
data = {
    "Name":["Tom","Nick","Xiomara","Astrid"],
    "Age":[22,20,19,25]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Seleccionamos la columnas
print(df["Age"])
```

```python
#Filtramos las filas
filter_colum = df[df["Age"]>19]
print(filter_colum)
```

```python
#Agregamos una columna
df["Profesion"]=["Piscologia","Administracion","Ingenieria","Medicina"]
print(df)
```

```python
#Obtener estadisticas basica
print(df.describe())
```

```python
#Ordenar datos Ascendente
sorted_df = df.sort_values(by="Age")
print(sorted_df)
```

```python
#Ordenar datos
sorted_df = df.sort_values(by="Age",ascending=False)
print(sorted_df)
```

```python
#Eliminar columna
df = df.drop(columns=["Age"])
print(df)
```
